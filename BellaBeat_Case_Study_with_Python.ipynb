## Consistent Date Formats

```python
## Changing the date format

# Sleep
sleep_df['SleepDay'] = pd.to_datetime(sleep_df['SleepDay'] )

sleep_df['new_date'] = [d.date() for d in sleep_df['SleepDay']]


# Calories
calories_df['ActivityHour'] = pd.to_datetime(calories_df['ActivityHour'])

calories_df['new_date'] = [d.date() for d in calories_df['ActivityHour']]
calories_df['new_time'] = [d.time() for d in calories_df['ActivityHour']]


# Intensities
intensities_df['ActivityHour'] = pd.to_datetime(intensities_df['ActivityHour'])

intensities_df['new_date'] = [d.date() for d in intensities_df['ActivityHour']]
intensities_df['new_time'] = [d.time() for d in intensities_df['ActivityHour']]


# Activity
activity_df['ActivityDate'] = pd.to_datetime(activity_df['ActivityDate'])

activity_df['new_date'] = [d.date() for d in activity_df['ActivityDate']]


# Weight
weight_df['Date'] = pd.to_datetime(weight_df['Date'])

weight_df['new_date'] = [d.date() for d in weight_df['Date']]
weight_df['new_time'] = [d.time() for d in weight_df['Date']]
```

## Combining data

Merging two data sets for better analysis of sleep hours and acitivities  per day on Date column.

```python
# merging sleep and activites data

combined_df = pd.merge(left =sleep_df, right= activity_df 
                            ,how="inner",on= ['Id','new_date']).drop(columns=['SleepDay','ActivityDate'])
```

###  Data Visualization

```python
fig = px.scatter(combined_df, x="TotalSteps", y="Calories",width=700, height=600, trendline='ols',trendline_scope="overall")
fig.update_layout(title= 'Total Steps vs. Calories', 
                  xaxis_title="Total Steps", yaxis_title="Calories")
fig.show();
```

Here we can see a fairly positive correaltion between "TotalSteps" and "Calories". 

```python
fig = px.scatter(combined_df, x="TotalMinutesAsleep", y="TotalTimeInBed",width=700, height=600, trendline='ols',trendline_scope="overall")
fig.show()
```

Here, we can see a srong positive corrrelation between the amount of time spent in bed and the minutes of sleep.

```python
intensities_new = intensities_df.groupby(by='new_time', dropna=True,as_index=False)
```

```python
intensities_new = intensities_df.groupby(by='new_time', dropna=True,as_index=False).mean()
```

```python
# Hourly Intensities
intensities_new
```

```python
fig = px.histogram(intensities_new, y='AverageIntensity',x='new_time',width=700, height=600)
fig.update_layout(title= 'Time vs. Average Total Intensity', 
                  xaxis_title="Time", yaxis_title="Average total intensity")
fig.show()
```

- Users are highly engage in activiy between 5pm - 8 pm.
- Least active between 11 pm - 5 am.

```python
fig = px.scatter(combined_df, x="TotalMinutesAsleep", y="SedentaryMinutes",width=700, height=600, trendline='lowess',trendline_scope="overall")
fig.update_layout(title= 'Total Minutes Asleep vs. Sedentary Minutes')
fig.show()
```

Total minutes asleep and Sedentary minutes have a mildly negative correlation, which means users might be able to increase their sleep minutes by reducing 'sedentary minutes' everyday.
